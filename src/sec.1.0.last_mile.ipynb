```python
from techminer2.database.search import ConcordantDocuments  # type: ignore
from techminer2.database.tools import RecordViewer #Â type: ignore


documents = []


for pattern in [
        "LAST_MILE",
        "ACCOMMODATE_OPTIMUM_DELIVERY_ORDERS",
        "A_TYPICAL_LAST_MILE_DELIVERY",
        "CONVENTIONAL_DELIVERY",
        "DELIVERIES",
        "DELIVERIES_INCREASES",
        "DELIVERY",
        "DRONE_DELIVERY",
        "EACH_DELIVERY",
        "EFFICIENT_LAST_MILE_DELIVERY",
        "LAST_MILE",
        "LAST_MILE_:_A_NOVEL",
        "LAST_MILE_DELIVERY",
        "LAST_MILE_DELIVERY_DRONES",
        "LAST_MILE_DELIVERY_ORDERS",
        "LAST_MILE_DELIVERY_RESOURCES",
        "LAST_MILE_LOGISTIC",
        "LAST_MILE_LOGISTICS",
        "LAST_MILE_LOGISTICS_PRESENT_UNIQUE_CHALLENGES",
        "LAST_MILE_PROBLEM",
        "LAST_MILE_PROBLEMS",
        "LMD",
        "LMDD",
        "SUSTAINABLE_LAST_MILE_DELIVERY",
        "THE_DELIVERIES",
        "THE_DELIVERY",
        "THE_LAST_MILE",
        "THE_LAST_MILE_DELIVERIES",
        "THE_LAST_MILE_DELIVERY",
        "THE_LAST_MILE_DELIVERY_PROBLEM",
        "THE_LAST_MILE_DELIVERY_PROBLEMS",
        "THE_LAST_MILE_LOGISTICS_PROBLEM",
        "THE_LAST_MILE_PROBLEM",
        "THE_LAST_MILE_PROBLEMS",
        "THE_LMD",
]:
    documents += (
        ConcordantDocuments()
        #
        .with_abstract_having_pattern(pattern)
        #
        .where_root_directory_is("../")
        .where_database_is("main")
        .where_record_years_range_is(None, None)
        .where_record_citations_range_is(None, None)
        .where_records_match(None)
        .where_records_ordered_by("date_newest")
    ).run()


ut = [doc.split("\n")[0] for doc in documents]



viewer = (
    RecordViewer()
    #
    .where_root_directory_is("../")
    .where_database_is("main")
    .where_record_years_range_is(None, None)
    .where_record_citations_range_is(None, None)
    .where_records_match(None)
    .where_records_ordered_by("date_newest")
)
documents = viewer.run()

documents = [
    doc
    for doc in documents
    if doc.split("\n")[0] in ut
]


with open("../reports/last_mile.txt", "w", encoding="utf-8") as f:
    for record in documents:
        f.write(f"{record}\n\n--\n\n")
        
```
