```python
from techminer2.science_mapping.citation import most_cited_documents

most_cited_documents(
    #
    # FUNCTION PARAMS:
    metric="global_citations",
    top_n=100,
    #
    # CHART PARAMS:
    title=None,
    field_label=None,
    metric_label=None,
    textfont_size=10,
    marker_size=7,
    line_width=1.5,
    yshift=4,
    #
    # DATABASE PARAMS:
    root_dir="./",
    database="main",
    year_filter=(2014, 2024),
    cited_by_filter=(None, None),
    document_type=[
        "Review",
    ],
).df_.to_csv("reports/literature_reviews.csv", sep="\t")
```
